```python
import pandas as pd

df = pd.read_csv("kaggle.csv")

# Q1 Probability of phones having 4G
prob_4g = df['four_g'].sum()/len(df)
print("Probability of 4G:",prob_4g)

# Q2 Probability of RAM greater than 3000
prob_ram = len(df[df['ram']>3000])/len(df)
print("Probability RAM >3000:",prob_ram)
```

```python
import pandas as pd
from scipy.stats import ttest_1samp

df=pd.read_csv("kaggle.csv")


# Q1 Test if average RAM is 2000
test=ttest_1samp(df['ram'],2000)

print(test)


# Q2 Check hypothesis result
if test.pvalue < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression


df=pd.read_csv("kaggle.csv")


# Feature and target
X=df[['battery_power']]
y=df['ram']


model=LinearRegression()


# Q1 Train model
model.fit(X,y)

print("Model trained successfully")


# Q2 Predict RAM for battery power 2000
new_phone = pd.DataFrame({
    'battery_power':[2000]
})

prediction=model.predict(new_phone)

print("Predicted RAM:", prediction[0])
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression


df=pd.read_csv("kaggle.csv")


# Create target
df['high_ram']=df['ram']>2000


X=df[['battery_power']]
y=df['high_ram']


model=LogisticRegression()


# Q1 Train classifier
model.fit(X,y)

print("Classifier trained successfully")


# Q2 Predict class
new_phone=pd.DataFrame({
    'battery_power':[2500]
})

prediction=model.predict(new_phone)

print("Prediction:", prediction[0])
```

```python
import pandas as pd
from sklearn.cluster import KMeans


df=pd.read_csv("kaggle.csv")


X=df[['ram','battery_power']]


model=KMeans(n_clusters=3)


# Q1 Create clusters
model.fit(X)

print(model.labels_)


# Q2 Display cluster centers
print(model.cluster_centers_)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


df=pd.read_csv("kaggle.csv")


df['target']=df['ram']>2000


X=df[['battery_power']]
y=df['target']


model=LogisticRegression()

model.fit(X,y)


prediction=model.predict(X)


# Q1 Calculate accuracy
print(accuracy_score(y,prediction))


# Q2 Display predictions
print(prediction)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV


df=pd.read_csv("kaggle.csv")


X=df[['ram','battery_power']]


model=KMeans()


params={
    'n_clusters':[2,3,4,5]
}


# Q1 Find best cluster number
grid=GridSearchCV(model,params)

grid.fit(X)


print(grid.best_params_)


# Q2 Display best model
print(grid.best_estimator_)
```
