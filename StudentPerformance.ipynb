```python
import pandas as pd
from scipy.stats import ttest_ind


df = pd.read_csv("student_performance_dataset.csv")



# Q1 Create groups

high_hours = df[
df["Study_Hours"]>5
]["Final_Score"]


low_hours = df[
df["Study_Hours"]<=5
]["Final_Score"]



# Q2 Perform t-test

t,p = ttest_ind(
high_hours,
low_hours
)


print("T value:",t)

print("P value:",p)



# Q3 Check hypothesis

if p < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")



# Q4 Group averages

print(
"High study average:",
high_hours.mean()
)

print(
"Low study average:",
low_hours.mean()
)
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error



df = pd.read_csv("student_performance_dataset.csv")



# Q1 Split data

X = df[["Study_Hours"]]

y = df["Final_Score"]


X_train,X_test,y_train,y_test = train_test_split(
X,y,test_size=0.2
)



model = LinearRegression()

model.fit(
X_train,
y_train
)


prediction = model.predict(
X_test
)


print("Predictions:")
print(prediction)



# Q2 Model error

error = mean_squared_error(
y_test,
prediction
)


print(
"Mean Squared Error:",
error
)
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


df = pd.read_csv("student_performance_dataset.csv")


# Q1 Predict Pass/Fail

X = df[
["Study_Hours",
"Attendance",
"Final_Score"]
]

y = df["Pass_Fail"]


X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.2
)


model = LogisticRegression()

model.fit(
    X_train,
    y_train
)


prediction = model.predict(
    X_test
)


print("Predictions:")
print(prediction)



# Q2 Accuracy

accuracy = accuracy_score(
    y_test,
    prediction
)


print(
"Accuracy:",
accuracy
)
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix


df = pd.read_csv("student_performance_dataset.csv")


# Q1 Train model

X = df[
["Study_Hours",
"Attendance",
"Final_Score"]
]

y = df["Pass_Fail"]


X_train,X_test,y_train,y_test = train_test_split(
X,y,test_size=0.2
)


model = LogisticRegression()

model.fit(
X_train,
y_train
)


prediction = model.predict(
X_test
)



print(
"Accuracy:",
accuracy_score(
y_test,
prediction
)
)



# Q2 Confusion Matrix

print(
confusion_matrix(
y_test,
prediction
)
)
```

```python
import pandas as pd

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier


df = pd.read_csv("student_performance_dataset.csv")


# Q1 Prepare data

X = df[
["Study_Hours",
"Attendance",
"Final_Score"]
]

y = df["Pass_Fail"]



model = KNeighborsClassifier()



parameters = {
"n_neighbors":[3,5,7,9]
}



grid = GridSearchCV(
model,
parameters,
cv=5
)


grid.fit(
X,y
)



# Q2 Best parameter

print(
"Best K value:",
grid.best_params_
)


print(
"Best Score:",
grid.best_score_
)
```
